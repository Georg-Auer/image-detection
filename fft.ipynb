```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import scipy.fftpack

# Number of samplepoints
N = 32

a = [6,11,16,21,26,31,36,41,46,51,56,61,66,71,76,81,86,91,96,101,106,111,116,121,126,131,136,141,146,151,156,161]
b = [10,15,20,25,30,35,40,45,50,55,60,65,70,75,80,100,130,135,130,100,70,40,45,85,110,140,145,140,110,80,50,55]
y1 = np.asarray(a)
y2 = np.asarray(b)

# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N)
#y = np.sin(55.0 * 2.0*np.pi*x) + 0.5*np.sin(80.0 * 2.0*np.pi*x)
yf1 = scipy.fftpack.fft(y1)
yf2 = scipy.fftpack.fft(y2)
xf = np.linspace(0.0, 1.0/(2.0*T), N/2)
# print(yf1)
# print(yf2)
fig, ax = plt.subplots()
ax.plot(xf, 2.0/N * np.abs(yf1[:N//2]), label="perfect loading")

ax.plot(xf, 2.0/N * np.abs(yf2[:N//2]), label="errornous loading")
plt.legend()
plt.show()
```
